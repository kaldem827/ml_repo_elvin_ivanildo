<h1>1.Number of grains of rice on the 100th day</h1>


```python
import matplotlib.pyplot as plt

days = 100
grains_per_day = []
grain_in_day = 1
total_rice_grains = 0

for day in range(days):
    grains_per_day.append(grain_in_day)
    grain_in_day = grain_in_day * 2

total_rice_grains = sum(grains_per_day)

print("Number of grains for 100 days:", total_rice_grains)

plt.figure(figsize=(10, 6))
plt.plot(range(1, days+1), grains_per_day, label='Grains Received on Day')
plt.xlabel('Day')
plt.ylabel('Number of Rice Grains')
plt.title('Change in Rice Grains Received')
plt.legend()
plt.grid(True)
plt.show()
```

<h1>2.Number of rice grains outside of the 100th day</h1>

```python
import datetime
def compute_sorori_shinzaemon(n_days=100):
    """
    A function that calculates the number of rice grains per day in the rice anecdote of Shinzaemon Sorori.

    Parameters
    ----------
    n_days : int
        Number of days to get rice (default: 100)

    Returns
    -------
    list_n_grains : list
        List of the number of rice grains received each day (list length equals n_days)
    list_total_grains : list
        List of the total number of rice grains received by a certain day (list length is equal to n_days)
    """

    # Initialize lists to store grains received each day and total grains received by a certain day
    list_n_grains = []
    list_total_grains = []

    # Initialize total grains received
    total_grains = 0

    # Loop through each day
    for day in range(1, n_days + 1):
        grains_on_day = 2 ** (day - 1)  # Calculate grains received on current day
        total_grains += grains_on_day  # Accumulate total grains received
        list_n_grains.append(grains_on_day)  # Append grains received on current day to list
        list_total_grains.append(total_grains)  # Append total grains received by current day to list

    return list_n_grains, list_total_grains



day_one = datetime.datetime(2023, 10, 2)
last_day = datetime.datetime(2024, 8, 31)
list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=(last_day - day_one).days)
print("Total rice grains till the end of the course: ", list_total_grains[-1])


plt.figure(figsize=(15, 6))
plt.plot(list_total_grains, label='Grains Received daily')
plt.xlabel('Day')
plt.ylabel('Number of Rice Grains')
plt.title('Change in Rice Grains Received')
plt.legend()
plt.grid(True)
plt.show()
```

<h1>3. How many people can live for how many days with the rice we get? </h1>

```python
def calculate_days_of_rice_sustenance(num_grains_of_rice: int, num_people: int) -> int:  
    average_grains_per_person = 12000 #average grains consumed per person according to some sources
    grains_per_day = average_grains_per_person * num_people #total daily consumption per day     
    number_of_days_to_live = num_grains_of_rice // grains_per_day #days of sustenance given the daily consumption
    return number_of_days_to_live

```

```python
#According to various sources, including the Food and Agriculture Organization (FAO), a commonly cited figure for daily rice consumption is around 400 grams per person per day in certain regions where rice is a staple food. 
#So based on the number, the daily consumption was calculated based on the number of people, then divided by the given total rice grains  
```
