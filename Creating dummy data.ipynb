<h1>1.Creating random numbers</h1>

```python
import numpy as np

mean = [-3, 0]

cov_matrix = [[1.0, 0.8],
              [0.8, 1.0]]

random_numbers_1 = np.random.multivariate_normal(mean, cov_matrix, 500)

print(random_numbers_1[:5])
```

<h1>2.Visualization by scatter plot</h1>

```python
import matplotlib.pyplot as plt
x = random_numbers_1[:, 0]
y = random_numbers_1[:, 1]

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.6, c='b', edgecolors='k')
plt.title('Scatter Plot of Random Numbers')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

<h1>3.Visualization by histogram</h1>

```python
plt.subplot(1, 2, 1)
plt.hist(x, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram for Dimension 1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.xlim(-6, 1)

plt.subplot(1, 2, 2)
plt.hist(y, bins=30, color='green', alpha=0.7, edgecolor='black')
plt.title('Histogram for Dimension 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.xlim(-4, 4)

plt.tight_layout()
plt.show()
```

<h1>4.Addition of data</h1>

```python
mean = [-3, 0]

cov_matrix = [[1.0, 0.8],
              [0.8, 1.0]]

random_numbers_1 = np.random.multivariate_normal(mean, cov_matrix, 500)

mean_2 = [0, -3]
cov_matrix_2 = [[1.0, 0.8], [0.8, 1.0]]

random_numbers_2 = np.random.multivariate_normal(mean_2, cov_matrix_2, 500)

x1, y1 = random_numbers_1[:, 0], random_numbers_1[:, 1]
x2, y2 = random_numbers_2[:, 0], random_numbers_2[:, 1]

plt.figure(figsize=(8, 6))
plt.scatter(x1, y1, alpha=0.6, c='b', edgecolors='k', label='Problem 1 (0)')
plt.scatter(x2, y2, alpha=0.6, c='r', edgecolors='k', label='Problem 4 (1)')
plt.title('Combined Scatter Plot of Problems 1 and 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

<h1>5.Data combination </h1>

```python
combined_data = np.vstack((random_numbers_1, random_numbers_2))

print(combined_data.shape)
```

<h1>6.Labeling</h1>

```python
labels_1 = np.zeros((500, 1))  # Label 0 for Problem 1
labels_2 = np.ones((500, 1))   # Label 1 for Problem 4

# Combine data and labels
labeled_data_1 = np.hstack((random_numbers_1, labels_1))
labeled_data_2 = np.hstack((random_numbers_2, labels_2))

# Combine both labeled datasets into a single ndarray
combined_labeled_data = np.vstack((labeled_data_1, labeled_data_2))

print(combined_labeled_data)
```
