<h1>4.Investigation of the iris dataset itself</h1>


The Iris dataset is one of the most well-known datasets in the machine learning community. It was introduced by the British biologist and statistician Ronald A. Fisher in his 1936 paper "The use of multiple measurements in taxonomic problems" as an example of discriminant analysis. The dataset provides a set of measurements for iris flowers from three different species, allowing for the classification of the species based on these measurements.

Each species is represented by 50 samples in the dataset. The primary objective is to classify the iris species based on the provided features.

Visualizations such as scatter plots of petal length vs. petal width or sepal length vs. sepal width can reveal clear distinctions between the species. Pair plots (also known as scatterplot matrices) show all possible scatter plots of feature pairs and are useful for identifying relationships and patterns in the data.

<h1>6.Creating a diagram</h1>


```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sr = df["Species"].value_counts()
labels = 'Setosa', 'Virgiclor','Virginca'
plt.title("A pie chart of the number of samples per label")
sr.plot.pie(autopct="%.1f%%", labels=labels);
```

```python
fig, axes = plt.subplots(1, 2)

sns.boxplot(x="Species", y="Petal_Length",data=df, ax=axes[0])
sns.boxplot(x="Species", y="Sepal length",data=df, ax=axes[1])

axes[0].set_xticklabels(['setosa','virgiclor','virginica'])
axes[0].set_title('Petal_Length boxplot')

axes[1].set_xticklabels(['setosa','virgiclor','virginica'])
axes[1].set_title('Sepal length boxplot')

fig.set_size_inches(14, 5)
plt.show()
     
```

```python
fig, axes = plt.subplots(1, 2)

sns.violinplot(x="Species", y="Petal_Length",data=df,ax=axes[0])
sns.violinplot(x="Species", y="Sepal length",data=df,ax=axes[1])

axes[0].set_xticklabels(['setosa','virgiclor','virginica'])
axes[0].set_title('sepal_length violinplot')

axes[1].set_xticklabels(['setosa','virgiclor','virginica'])
axes[1].set_title('Petal_Length violinplot')

fig.set_size_inches(14, 5)
plt.show()
     
```

 Both box plots and violin plots are valuable tools for visualizing data distributions, with box plots offering a simpler, more straightforward summary and violin plots providing a richer, more detailed view of the data distribution. 

<h1>7.Confirming the relationship between features</h1>


```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

fig.suptitle('- Create scatter plots of all combinations (6) such that 2 out of 4 features are selected by color.')

sns.scatterplot(ax=ax[0,0], data=df, x='Sepal length', y='Sepal width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,1], data=df, x='Sepal length', y='Petal_Length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,2], data=df, x='Sepal length', y='Petal_Width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,0], data=df, x='Sepal width', y='Petal_Length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,1], data=df, x='Sepal width', y='Petal_Width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,2], data=df, x='Petal_Length', y='Petal_Width', hue='Species', palette="deep")

sns.set()
     
```

```python

rel=sns.pairplot(df, hue='Species')

rel.fig.subplots_adjust(top=.9)
rel.fig.suptitle('- Create a scatterplot matrix that displays all feature combinations at once.')

sns.set()
```
