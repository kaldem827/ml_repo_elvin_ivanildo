<h1>1. Linear function </h1>

```python
import numpy as np
x = np.arange(-50, 50.0, 0.1)# fill the array with values between -50 50 with a 0.1 step

y = (1/2) * x + 1# f(x) or y that maps to x 

print(y)
```

<h1>2. Array combination </h1>

```python
data = np.column_stack((x, y))
print(data)
```

<h1>3.Find the gradient</h1>

```python
x_slice_one = x[1:]
x_slice_two = x[:-1]

y_slice_one = y[1:]
y_slice_two = y[:-1]
change_in_x = x_slice_one - x_slice_two
change_in_y = y_slice_one - y_slice_two
gradient = change_in_y / change_in_x
print(gradient)
```

<h1>4. Draw a graph </h1>

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))

#  Linear function
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Linear Function: y = 0.5x + 1')
plt.title('Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Plot 2: Gradient
plt.subplot(1, 2, 2)
# x[:-1] to arrange x to gradient size
plt.plot(x[:-1], gradient, label='Gradient')
plt.title('Gradient')
plt.xlabel('x')
plt.ylabel('Gradient')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

<h1>5. Python functionalization </h1>

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    def gradient(x, y):#helper function to calculate the gradient
        sl_x_one = x[1:]
        sl_x_two = x[:-1]
        sl_y_one = y[1:]
        sl_y_two = y[:-1]
       
        ch_x = sl_x_two - sl_x_one
        ch_y = sl_y_two - sl_y_one
        
        return ch_y / ch_x
    x = np.arange(*x_range)
    y = function(x)
    
    array_xy = np.column_stack((x, y))
    
    grad = gradient(x, y)
    
    return array_xy, grad

def func1(x):
    return x**2
def func2(x):
    return  2*x**2 + 2**x
def func3(x):
    return np.sin(np.sqrt(x))

array_one, gradient_one = compute_gradient(func1)
array_two, gradient_two = compute_gradient(func2)
array_three, gradient_three = compute_gradient(func3, (0, 50.1, 0.1))

# Plot for function 1
plt.subplot(1, 2, 1)
plt.plot(array_one[:, 0], array_one[:, 1], label='Function 1: x^2')
plt.title('Function 1: x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()

plt.show()

# Plot for function 2
plt.subplot(1, 2, 1)
plt.plot(array_two[:, 0], array_two[:, 1], label='Function 2: 2x^2 + 2x')
plt.title('Function 2: 2x^2 + 2x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.tight_layout()

plt.show()

# Plot for function 3
plt.subplot(1, 2, 1)
plt.plot(array_three[:, 0], array_three[:, 1], label='Function 3: sin(sqrt(x))')
plt.title('Function 3: sin(sqrt(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.tight_layout()
plt.show()
```

<h1>6. Find the minimum value </h1>

```python

# Find the minimum value of y and its index
min_y = np.min(array_three[:, 1])
min_y_index = np.argmin(array_three[:, 1])

# Display the minimum value of y and its corresponding x
print("Minimum value of y:", min_y)
print("Corresponding x value at the minimum y: ", array_three[min_y_index, 0])

# slop before and after min
slope_before = gradient_three[min_y_index - 1]  # slope before the minimum y
slope_after = gradient_three[min_y_index]  # slope after the minimum y

print("Slope before taking the minimum y:", slope_before)
print("Slope after taking the minimum y:", slope_after)
```
